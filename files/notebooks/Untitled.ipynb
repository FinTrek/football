```python
import pandas as pd
import numpy as np
import sys
import os
import json

path_project = "/home/igorcosta/PycharmProjects/football/"
sys.path.insert(1, path_project)
from core.config_loader import SysConfig
```

```python
exp1 = ['be_btts','total','performance']
exp2 = ['be_btts_odds','market','market']
exp3 = ['be_btts_odds','total','total']
exps = [exp1, exp2, exp3]

frame = list()
for c in ['GaussianNB','LogisticRegression','XGBClassifier', 'mkc']:
    for e in exps:
        for ch in ['Brazil A', 'Brazil B', 'France A', 
                   'Portugal A', 'Italy A', 'Netherlands A', 'Spain A', 'Germany A', 'England A',
                   'total']:
            
            filename = ch + '.json'
            score_dir = os.path.join(SysConfig.path('results'), c, e[0], e[1])
            score_file = os.path.join(score_dir, filename)
            
            with open(score_file, 'r') as outfile:
                data = outfile.read()
                res = json.loads(data)
            frame.append([c, e[2], ch, res['acc_test'], res['rps_test']])
            
resume = pd.DataFrame(frame, columns=['classifier','features', 'championship', 'acc','rps'])
```

```python
for ch in ['Brazil A', 'Brazil B', 'France A', 
           'Portugal A', 'Italy A', 'Netherlands A', 'Spain A', 'Germany A', 'England A',
           'total']:
    temp = resume[resume['championship'] == ch]
    temp = temp.sort_values('rps', ascending=False)
    print(temp)
    print("----")
```
