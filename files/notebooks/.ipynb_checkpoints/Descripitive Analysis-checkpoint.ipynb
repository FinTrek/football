```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import sys
import os
path_project = '/home/igorcosta/PycharmProjects/football/'
sys.path.insert(1, path_project)

from core.config_loader import Config
datasets_path = Config.path('datasets')
plt.style.use('seaborn-whitegrid')
```

```python
dataset_id = 'sm_accum_odds'
file_path = os.path.join(Config.path('datasets'), dataset_id, 'final.csv')
df = pd.read_csv(file_path)
```

```python
df = df[df['odds_p_mean_home'].notna()]
df = df[df['odds_p_mean_draw'].notna()]
df = df[df['odds_p_mean_away'].notna()]
```

```python
trends_list = ['goals', 'corners', 'on_target', 'off_target', 'attacks', 'dangerous_attacks']
cards_list = ['yellow_cards', 'red_cards']
ratio_list = ['possession']
total_list = trends_list + cards_list + ratio_list
```

```python
df['result_t'] = np.where(df['result']=='H', 1, np.where(df['result'] == 'D', 2, 3))
```

```python
pred = pd.DataFrame()
pred['id'] = df.index
pred.index = pred['id']
pred['result_t'] = df['result_t']

hits = pd.DataFrame()
hits['id'] = df.index
hits.index = hits['id']
hits['result_t'] = df['result_t']

for t in trends_list:
    for i in range(0,96):
        col_home = '_'.join(['accum_trends',t,'home', str(i)])
        col_away = '_'.join(['accum_trends',t,'away', str(i)])
        col = '_'.join([t,str(i)])
        pred[col] = np.where(df[col_home]>df[col_away],1, np.where(df[col_home]==df[col_away], 2, 3))
        hits[col] = pred[col] == pred['result_t']
        

for t in cards_list:
    for i in range(0,96):
        col_home = '_'.join(['accum_cards',t,'home', str(i)])
        col_away = '_'.join(['accum_cards',t,'away', str(i)])
        col = '_'.join([t,str(i)])
        pred[col] = np.where(df[col_home]<df[col_away],1, np.where(df[col_home]==df[col_away], 2, 3))
        hits[col] = pred[col] == pred['result_t']
        
for t in ratio_list:
    for i in range(0,96):
        col_home = '_'.join(['ratio_trends',t, str(i)])
        col = '_'.join([t,str(i)])
        pred[col] = np.where(df[col_home]>0.5,1, np.where(df[col_home]==0.5, 2, 3))
        hits[col] = pred[col] == pred['result_t']
        
pred_normal = pred
hits_normal = hits
```

```python
pred = pd.DataFrame()
pred['id'] = df.index
pred.index = pred['id']
pred['result_t'] = df['result_t']

hits = pd.DataFrame()
hits['id'] = df.index
hits.index = hits['id']
hits['result_t'] = df['result_t']

for t in trends_list:
    for i in range(0,96):
        col_home = '_'.join(['accum_trends',t,'home', str(i)])
        col_away = '_'.join(['accum_trends',t,'away', str(i)])
        col = '_'.join([t,str(i)])
        pred[col] = np.where(df[col_home]>=df[col_away],1,3)
        hits[col] = pred[col] == pred['result_t']
        
for t in cards_list:
    for i in range(0,96):
        col_home = '_'.join(['accum_cards',t,'home', str(i)])
        col_away = '_'.join(['accum_cards',t,'away', str(i)])
        col = '_'.join([t,str(i)])
        pred[col] = np.where(df[col_home]<df[col_away],1,3)
        hits[col] = pred[col] == pred['result_t']
        
for t in ratio_list:
    for i in range(0,96):
        col_home = '_'.join(['ratio_trends',t, str(i)])
        col = '_'.join([t,str(i)])
        pred[col] = np.where(df[col_home]>=0.5,1,3)
        hits[col] = pred[col] == pred['result_t']
        
pred_hf = pred
hits_hf = hits
```

```python
def get_resume(df_hits):
    
    resume = pd.DataFrame()
    total_list = trends_list + cards_list + ratio_list
    for t in total_list:
        hits = list()
        for i in range(0,96):
            col_hit = '_'.join([t,str(i)])
            resume.loc[i,t] = len(df_hits[df_hits[col_hit] == True])/len(df_hits)
    return resume

resume_normal = get_resume(hits_normal)
resume_hf = get_resume(hits_hf)
```

```python
def count_result(df_count, row):
    if row['result'] == 'H':
        df_count['h'] += 1
    elif row['result'] == 'D':
        df_count['d'] += 1
    elif row['result'] == 'A':
        df_count['a'] += 1
```

```python
try:
    df_exp1 = pd.read_csv('df_exp1.csv')
except:
    df_exp1 = pd.DataFrame()
    
    for j in range(0,96):

        print(j, end=' ')
        home = 0
        away = 0

        home_first = {'h': 0, 'd': 0, 'a': 0}
        away_first = {'h': 0, 'd': 0, 'a': 0}

        col_home = '_'.join(['accum_trends_goals_home', str(j)])
        col_away = '_'.join(['accum_trends_goals_away', str(j)])
        exp = df[(df[col_home] == 1) & (df[col_away] == 1)]

        for row_tuple in exp.iterrows():
            row = row_tuple[1]
            for i in range(0,j):
                col_home_i = '_'.join(['accum','trends','goals', 'home', str(i)])
                col_away_i = '_'.join(['accum','trends','goals', 'away', str(i)])
                if row[col_home_i] == 1:
                    home +=1
                    count_result(home_first, row)
                    break
                elif row[col_away_i] == 1:
                    away +=1
                    count_result(away_first, row)
                    break

        first_win = home_first['h'] + away_first['a']
        draw = home_first['d'] + away_first['d']
        second_win = home_first['a'] + away_first['h']
        p_value = round(stats.binom_test(first_win, first_win+second_win, p=0.5), 4)
        p_value_home = round(stats.binom_test(home_first['h'], home_first['h']+home_first['a'], p=0.5), 4)
        p_value_away = round(stats.binom_test(away_first['a'], away_first['h']+away_first['a'], p=0.5), 4)

        df_exp1.loc[j,'total'] = len(exp)

        df_exp1.loc[j,'first_win'] = first_win
        df_exp1.loc[j,'draw'] = draw
        df_exp1.loc[j,'second_win'] = second_win
        df_exp1.loc[j,'p_value'] = p_value

        df_exp1.loc[j,'h_first_win'] = home_first['h']
        df_exp1.loc[j,'h_draw'] = home_first['d']
        df_exp1.loc[j,'h_second_win'] = home_first['a']
        df_exp1.loc[j,'h_p_value'] = p_value_home

        df_exp1.loc[j,'a_first_win'] = away_first['a']
        df_exp1.loc[j,'a_draw'] = away_first['d']
        df_exp1.loc[j,'a_second_win'] = away_first['h']
        df_exp1.loc[j,'a_p_value'] = p_value_away
        
    df_exp1.to_csv('df_exp1.csv')
```

```python
df_exp2 = pd.DataFrame()

for j in range(60,96):
    
    print(j)
    home = {'turn':0 , 'balance': 0, 'reaction': 0}
    away = {'turn':0 , 'balance': 0, 'reaction': 0}

    home_winning_turn = {'h': 0, 'd': 0, 'a': 0}
    home_winning_balance = {'h': 0, 'd': 0, 'a': 0}
    home_winning_reaction = {'h': 0, 'd': 0, 'a': 0}

    away_winning_turn = {'h': 0, 'd': 0, 'a': 0}
    away_winning_balance = {'h': 0, 'd': 0, 'a': 0}
    away_winning_reaction = {'h': 0, 'd': 0, 'a': 0}

    col_home = '_'.join(['accum','trends','goals', 'home', str(j)])
    col_away = '_'.join(['accum','trends','goals', 'away', str(j)])
    exp2a = df[(df[col_home] == 2) & (df[col_away] == 1)]
    exp2b = df[(df[col_home] == 1) & (df[col_away] == 2)]
    
    for row_tuple in exp2a.iterrows():
        
        row = row_tuple[1]
        for i in range(0,j):
            col_home = '_'.join(['accum','trends','goals', 'home', str(i)])
            col_away = '_'.join(['accum','trends','goals', 'away', str(i)])

            if row[col_home] == 1 and row[col_away] == 1:
                home['balance'] += 1
                count_result(home_winning_balance, row)
                break

            if row[col_home] == 0 and row[col_away] == 1:
                home['turn'] += 1
                count_result(home_winning_turn, row)
                break

            if row[col_home] == 2 and row[col_away] == 0:
                home['reaction'] += 1
                count_result(home_winning_reaction, row)
                break


    for row_tuple in exp2b.iterrows():

        row = row_tuple[1]
        for i in range(0,60):

            col_home = '_'.join(['accum','trends','goals', 'home', str(i)])
            col_away = '_'.join(['accum','trends','goals', 'away', str(i)])

            if row[col_home] == 1 and row[col_away] == 1:
                away['balance'] += 1
                if row['result'] == 'H':
                    away_winning_balance['h'] += 1
                elif row['result'] == 'D':
                    away_winning_balance['d'] += 1
                elif row['result'] == 'A':
                    away_winning_balance['a'] += 1
                break

            if row[col_home] == 1 and row[col_away] == 0:
                away['turn'] += 1
                if row['result'] == 'H':
                    away_winning_turn['h'] += 1
                elif row['result'] == 'D':
                    away_winning_turn['d'] += 1
                elif row['result'] == 'A':
                    away_winning_turn['a'] += 1
                break

            if row[col_home] == 0 and row[col_away] == 2:
                away['reaction'] += 1
                if row['result'] == 'H':
                    away_winning_reaction['h'] += 1
                elif row['result'] == 'D':
                    away_winning_reaction['d'] += 1
                elif row['result'] == 'A':
                    away_winning_reaction['a'] += 1
                break

    turn_total = home['turn'] + away['turn']
    turn_winning = home_winning_turn['h'] + away_winning_turn['a']
    turn_draw = home_winning_turn['d'] + away_winning_turn['d']
    turn_loosing = home_winning_turn['a'] + away_winning_turn['h']
    
    balance_total = home['balance'] + away['balance']
    balance_winning = home_winning_balance['h'] + away_winning_balance['a']
    balance_draw = home_winning_balance['d'] + away_winning_balance['d']
    balance_loosing = home_winning_balance['a'] + away_winning_balance['h']
    
    reaction_total = home['reaction'] + away['reaction']
    reaction_winning = home_winning_reaction['h'] + away_winning_reaction['a']
    reaction_draw = home_winning_reaction['d'] + away_winning_reaction['d']
    reaction_loosing = home_winning_reaction['a'] + away_winning_reaction['h']
    
    pop_balance = balance_winning*[0] + balance_draw*[1] + balance_loosing*[2]
    pop_turn = turn_winning*[0] + turn_draw*[1] + turn_loosing*[2]
    pop_reaction = reaction_winning*[0] + reaction_draw*[1] + reaction_loosing*[2]

    p_balance_reaction = stats.ttest_ind(pop_balance, pop_reaction)
    p_balance_turn = stats.ttest_ind(pop_balance, pop_turn)
    p_turn_reaction = stats.ttest_ind(pop_turn, pop_reaction)
    
    df_exp2.loc[j,'turn_total'] = turn_total
    df_exp2.loc[j,'turn_winning'] = turn_winning
    df_exp2.loc[j,'turn_draw'] = turn_draw
    df_exp2.loc[j,'turn_loosing'] = turn_loosing
    
    df_exp2.loc[j,'balance_total'] = balance_total
    df_exp2.loc[j,'balance_winning'] = balance_winning
    df_exp2.loc[j,'balance_draw'] = balance_draw
    df_exp2.loc[j,'balance_loosing'] = balance_loosing
    
    df_exp2.loc[j,'reaction_total'] = reaction_total
    df_exp2.loc[j,'reaction_winning'] = reaction_winning
    df_exp2.loc[j,'reaction_draw'] = reaction_draw
    df_exp2.loc[j,'reaction_loosing'] = reaction_loosing
    
    df_exp2.loc[j,'p_balance_reaction'] = p_balance_reaction.pvalue
    df_exp2.loc[j,'p_balance_turn'] = p_balance_turn.pvalue
    df_exp2.loc[j,'p_turn_reaction'] = p_turn_reaction.pvalue
```

```python
def get_sum(df_home, df_away):
    
    remains = df_home['h'] + df_away['a']
    draw = df_home['d'] + df_away['d']
    turn = df_home['a'] + df_away['h']
    
    return remains, draw, turn

draw_remains, draw_draw, draw_turn = get_sum(home_winning_draw, away_winning_draw)
turn_remains, turn_draw, turn_turn = get_sum(home_winning_turn, away_winning_turn)
reaction_remains, reaction_draw, reaction_turn = get_sum(home_winning_reaction, away_winning_reaction)

print(draw_remains, draw_draw, draw_turn)
print(turn_remains, turn_draw, turn_turn)
print(reaction_remains, reaction_draw, reaction_turn)
```

```python
pop1 = draw_remains*[0] + draw_draw*[1] + draw_turn*[2]
pop2 = turn_remains*[0] + turn_draw*[1] + turn_turn*[2]
pop3 = reaction_remains*[0] + reaction_draw*[1] + reaction_turn*[2]

print(stats.ttest_ind(pop1, pop2))
print(stats.ttest_ind(pop1, pop3))
print(stats.ttest_ind(pop2, pop3))
```
